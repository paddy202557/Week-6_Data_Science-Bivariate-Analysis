# 10.What is the probability Density Function of the salary range from 700000 to 900000?

```python
#The PDF provides a way to calculate the probability that a random variable falls within a specific interval.
```

```python
from QQUnivariate import QQUnivariate
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv("PrePlacement.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
#A Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable 
# taking on a particular value within a given range. 
#A probability function that describes the density of any continuous random variable within a specific range of values is
# called a probability density function or probability distribution function.
```

```python
def get_pdf_probability_forloop(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    #Normal Distribution: scipy.stats.norm
    
    # Draw the graph using matplotlib
    pyplot.title("Probability Distribution for Salary Range")
    pyplot.ylabel("Probability Density")
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    pyplot.legend()
    pyplot.show()
    
    
    # Create a normal distribution object
    # normal distribution -bell-shaped curve and is defined by two parameters:the mean (mu) & the standard deviation (sigma)
    
    sample=dataset
    #Calculate parameters
    sample_mean=sample.mean()
    sample_std= sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    
    #Calculate the normal distribution
    dist = norm(sample_mean, sample_std)
    
    # Calculate pdf for a rnge startrange, endrange
    
    Values= []
    for values in range(startrange, endrange):
        Values.append(values)
        
    # pdf: density of any continuous random variable within a specific range of values    
    probabiliites=[dist.pdf(value) for value in Values]
    prob=sum(probabiliites)
    
    print ("The area between range",(startrange,endrange,sum(probabiliites)))
    return prob 
```

```python
get_pdf_probability_forloop(dataset["salary"],700000,900000)
```

# 13 Plot any useful graph and explain it

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("PrePlacement.csv")
```

```python
dataset.drop(columns=["sl_no"],inplace=True, errors='ignore')
```

```python
dataset
```

```python
sns.heatmap(dataset.corr(),annot=True, cmap="coolwarm")
plt.title("Coorelation Heatmap of PrePlacement Data Set")
plt.show()
#This heatmap displays the correlation between different features in the PrePlacement dataset,
# with higher correlation values represented by darker colors.
```

```python
sns.displot(data=dataset, x="salary", hue="gender",multiple="stack")
plt.title("Male & Female Vs Salary")
plt.show()
```

```python
sns.displot(data=dataset, x="salary", hue="specialisation",multiple="stack")
plt.title("Specialisation Vs Salary")
plt.show()
```
