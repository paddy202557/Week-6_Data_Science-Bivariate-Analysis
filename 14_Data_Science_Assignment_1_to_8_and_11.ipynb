# 4.Find the relationship between "salary" & "mba_p"

###### Find the covariance between salary & mba_p columns

```python
# Covariance calculates the average product of deviations from mean.
#
# Covariance: Variance of two variables (linear relationship between two variables) X & Y together is Covariance 
#    (How they change or  do not change together). If both the values are increasing and directly proportional 
#    to each other then it is a positive covariance.If one value is increasing and the other value is decreasing then 
#    it is a negative covariance.
```

```python
#Covariance is a statistical measure that indicates the extent to which two random variables change together. 
#  It quantifies the joint variability of the variables, showing whether they tend to increase or decrease in tandem. 
#  A positive covariance indicates that the variables move in the same direction, while a negative covariance indicates 
#  they move in opposite directions. Covariance is similar to variance, but while variance measures the variability 
#  of a single variable, covariance measures how two variables vary together.
```

```python
dataset.cov()
```

Result salary increases and  decreases  with mba_p positively and negatively

## 7.Test the Analysis of Variance between "etest_p" & "mba_p" at signifance level 5%.
(Make decision using Hypothesis Testing)

###### Hypothesis : A premise or claim that we want to test
Null Hypothesis-  Ho (default  hypothesis or  currently accepted hypothesis or currently accepted value for a parameter)

Alternative Hypothesis- Ha-  Also called Research hypothesis, Involves a claim to be tested

###### Unpaired "ttest_ind" Independent test

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
data_etest_p = dataset['etest_p'] 
data_mba_p = dataset['mba_p']
t_stat, p_value = ttest_ind(data_etest_p,data_mba_p)
ttest_ind(data_etest_p,data_mba_p)
```

```python
#Analysis Result
if p_value < 5:
    print("Reject the null hypothesis: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

# 8.Test the similarity between the degree_t (Sci&Tech) & Special(Mkt&HR) 
with respect to salary at significance level of 5% (Makedecision using Hypothesis Testing)


###### Hypothesis : A premise or claim that we want to test
Null Hypothesis-  Ho (default  hypothesis or  currently accepted hypothesis or currently accepted value for a parameter)

Alternative Hypothesis- Ha-  Also called Research hypothesis, Involves a claim to be tested

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
data_degree_t = dataset[dataset['degree_t']=='Sci&Tech']['salary']
data_special = dataset[dataset['specialisation']=='Mkt&HR']['salary']
t_stat, p_value = ttest_ind(data_degree_t ,data_special)
ttest_ind(data_degree_t ,data_special)
```

```python
#Analysis Result
if p_value < 5:
    print("Reject the null hypothesis: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```
